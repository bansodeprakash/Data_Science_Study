```python
# Split route into separate segments
segments = df['Route'].str.split(' → ', expand=True)

# Get the maximum number of segments in any route
max_segments = segments.shape[1]

# Create column names for the route segments
segment_cols = [f"route_segment{i+1}" for i in range(max_segments)]

# Assign the split segments to new columns
segments.columns = segment_cols

# Concatenate the split segments with the original dataset
df_encoded = pd.concat([df, segments], axis=1)

# Convert categorical variables into numerical representations using one-hot encoding
encoded_segments = pd.get_dummies(df_encoded[segment_cols])

# Merge the encoded columns with the original dataset
df_encoded = pd.concat([df_encoded, encoded_segments], axis=1)

# Remove the original categorical columns and segment columns
df_encoded.drop(['Route'] + segment_cols, axis=1, inplace=True)

# Now you can use the df_encoded dataset for your ML model
```

```python
def duration_to_hours(duration):
    hours = 0
    minutes = 0
    if 'h' in duration:
        hours = int(duration.split('h')[0])
    if 'm' in duration:
        minutes = int(duration.split('m')[0].split()[-1])
    return hours + minutes / 60

df1['Duration_hours'] = df1['Duration'].apply(duration_to_hours)
```
